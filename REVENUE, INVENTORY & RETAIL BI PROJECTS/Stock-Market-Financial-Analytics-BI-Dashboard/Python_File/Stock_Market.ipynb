````python
'''
Explanation of `rolling()` and `pct_change()` functions:

1.  `rolling()` function:
    *   **Purpose**: The `rolling()` function provides rolling window calculations. It creates a 'rolling window object' over a specified window size (e.g., 50 days, 200 days). Once this object is created, you can apply various aggregation functions (like `mean()`, `sum()`, `std()`, `min()`, `max()`) to the data within each window.
    *   **How it was used in the notebook**: In the notebook, `df.Close.rolling(50).mean()` and `df.Close.rolling(200).mean()` were used to calculate the 50-day and 200-day Simple Moving Averages (SMAs) of the 'Close' price for each stock. These moving averages help smooth out price data to identify trends.
    *   **Example**: If you have daily stock prices, `df.Close.rolling(50).mean()` calculates the average closing price over the past 50 trading days for each point in time.

2.  `pct_change()` function:
    *   **Purpose**: The `pct_change()` function calculates the percentage change between the current and a prior element in a Series or DataFrame. It's commonly used to compute returns for financial data.
    *   **How it was used in the notebook**: `df.Close.pct_change()` was used to calculate the daily percentage change in the 'Close' price for each stock. This is a fundamental metric for understanding daily stock returns.
    *   **Example**: For a stock price sequence [10, 11, 10.5], the `pct_change()` would yield `[NaN, (11-10)/10, (10.5-11)/11]`, which is `[NaN, 0.1, -0.04545]`. The `NaN` for the first entry is because there is no prior value to compare it against.

In summary, `rolling()` is for window-based computations (like moving averages), and `pct_change()` is for calculating percentage differences between consecutive observations.
'''
````

### Common Pandas Time-Series Functions

Pandas is exceptionally well-suited for time-series analysis due to its `DatetimeIndex` and powerful functions. Here are some key functionalities:

1.  **`pd.to_datetime()`**: Converts a column to datetime objects, which is crucial for any time-series operation. We've already seen a 'Date' column, but it's important to ensure its `dtype` is `datetime`.
    *   **Example**: `df['Date'] = pd.to_datetime(df['Date'])`

2.  **`df.set_index('Date', inplace=True)`**: Sets a datetime column as the DataFrame's index. This allows for powerful time-based indexing and slicing.
    *   **Example**: `df_apple.set_index('Date', inplace=True)`

3.  **Time-based Indexing and Slicing**: Once you have a `DatetimeIndex`, you can easily select data by date, month, year, or a range.
    *   **Examples**:
        *   `df_apple.loc['2015']` (select all data for 2015)
        *   `df_apple.loc['2019-01']` (select all data for January 2019)
        *   `df_apple.loc['2018-06-01':'2018-06-30']` (select a specific date range)

4.  **`resample()`**: This is incredibly powerful for changing the frequency of your time series (e.g., from daily to weekly, monthly, or yearly) and applying aggregation functions.
    *   **Example (resampling daily data to monthly mean close price)**:
        ```python
        df_apple['Close'].resample('M').mean()
        ```
        Common frequency aliases:
        *   `'D'` (daily)
        *   `'W'` (weekly)
        *   `'M'` (monthly)
        *   `'Q'` (quarterly)
        *   `'A'` (yearly)

5.  **`rolling()`**: As we've already discussed, this function creates a rolling window object to perform operations like moving averages.
    *   **Example**: `df['Close'].rolling(window=30).mean()` (30-day rolling mean)

6.  **`expanding()`**: Similar to `rolling()`, but the window expands to include all data from the beginning up to the current point.
    *   **Example**: `df['Close'].expanding().mean()` (cumulative mean)

7.  **`diff()`**: Calculates the difference between the current and a prior element. Useful for calculating absolute changes over time.
    *   **Example**: `df['Close'].diff()` (daily change in closing price)

8.  **`shift()`**: Shifts data by a specified number of periods. Useful for comparing current values with past or future values.
    *   **Example**: `df['Close'].shift(1)` (previous day's closing price, which we used)

9.  **Time Zone Handling**: Pandas allows you to localize time-series data to specific time zones and convert between them.
    *   **Example**:
        ```python
        # Localize to 'US/Eastern'
        df.index = df.index.tz_localize('US/Eastern')
        # Convert to 'Europe/London'
        df.index = df.index.tz_convert('Europe/London')
        ```

These functions, especially when combined, allow for comprehensive analysis and manipulation of time-series datasets in Pandas.
